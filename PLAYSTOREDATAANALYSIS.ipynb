- #### DATA CLEANING
     - ##### cleaning the numerical columns with character

  - ##### #Remove text units and convert to numbers

```python
df.head()  #displaying the table to check the columns with error or needs cleaning
```

```python
df['Size']
```

```python
df['Size'].value_counts()
```

```python
df['Installs'] = pd.to_numeric(
    df['Installs']
    .astype(str)
    .str.replace(r'[+,]', '', regex=True),  # remove + and commas
    errors='coerce'
)
df
```

- #### after cleaning  we can see in the above table that the '+' in the install column as removed 

- ##### now we can check after cleaning does the type and info changed

```python
df.info()
```

- ##### In the same way we can also change the other column and others needing changes 
- ##### we can see that in the price column dollar symbol $ is there so we are converting this to integer column by removing it. 

```python
df['Reviews']=pd.to_numeric(
    df['Reviews']
    .astype(str)
    .str.replace(r'[^0-9,-]','',regex=True),
    errors='coerce'
)
df
```

```python
df['Price']=pd.to_numeric(
    df['Price']
    .astype(str)
    .str.replace(r'[^0-9,-]','',regex=True),
    errors='coerce'
)
df
```

```python
df.sample(15)
#checking whether the price column changed   
```

- ### How to handle data
     ##### For example consider a data with 1000 rows and 8 columns 
     ##### A- 900 missing -Since a lot of the values are missing in consideration with the total number of rows is 1000  in this column so we can drop this column(Also consider of the importance of the column in analysis before droping )
     ##### B-12 :- fill if data is less or drop rows
     ##### C- 0 :- No missing values
     ##### D- 200 :- drop the row with columns having missing values if it is not relevant
     

- #### How to fill missing data
   - ##### Numerical:- Fill with median
   - ##### Categoriical:- Fill with mode

- ##### Filling With Median:- 

```python
# Filling With Median:- 
df['Rating'].fillna(df['Rating'].median(),inplace=True)
df.isnull().sum()
```

- ### Duplicate App Entries
    - Total apps: 10,354 
    - Unique apps: 9,656
    - 700 duplicate entries exist
    - Inference:Dataset likely contains same apps repeated (maybe different versions or errors)
    - Can affect accuracy of analysis like installs, ratings, etc.
      
- ### Category Concentration
    - Total categories: 33
    - Most dominant: FAMILY (1,942 apps)
    - Inference:Play Store is heavily concentrated in FAMILY category
    - Indicates that High demand for general-purpose / kid-friendly apps
    - Possible high competition in this category
- ### Free Apps Dominate Market
    - Free apps: 9,590 out of 10,354 (~93%)
    - Inference:Market strongly favors free apps over paid
    - Revenue model likely depends on Advertisements and In-app purchases
    - Paid apps form a very small segment
- ### Target Audience is Broad
    - Most common rating: Everyone (8,380 apps)
    - Inference:Developers target maximum audience reach
    - Apps are designed to be:Safe and Widely usable
    - Less focus on niche/age-restricted apps
- ### Tools Genre is Most Popular
    - Top genre: Tools (842 apps)
    - Total genres: 119
    - Inference:Utility apps (cleaners, file managers, etc.) are very common
    - High competition in Tools category
    - Indicates user demand for practical apps

- ###  Android Compatibility Strategy
    - Most common version: 4.1 and up (2,379 apps)
    - Inference:Developers prioritize backward compatibility
    - Apps are built to support:Older devices
    - Helps increase installs, especially in developing regions

- ### High Diversity but Uneven Distribution
    - Genres: 119 unique
    - Categories: 33 unique
    - Inference:Play Store has wide variety of apps,But usage and development are concentrated in few dominant areas


- ### Repetition of Popular Apps
    - Most frequent app: ROBLOX (9 times)
    - Inference:Popular apps may appear multiple times (versions, updates)
- ## CONCLUSION
    - ### The Play Store ecosystem is:

       - Free-dominated
       - Mass audience focused
       - Highly competitive in common categories (Family, Tools) 
       - Designed for accessibility (older Android versions)
